```python
import threading
import time
import random
import signal


shared_resource = 0
lock = threading.Lock()


stop_flag = False


def task(name):
    global shared_resource


    lock.acquire()
    try:
        print(f"Thread {name} acquired the lock")
        time.sleep(random.randint(1, 3))
        shared_resource += 1
        print(f"Thread {name} released the lock")

    finally:

        lock.release()


def worker(name):
    print(f"Thread {name} started")
    for i in range(5):
        task(name)
    print(f"Thread {name} finished")


def stop_monitor():
    global stop_flag
    while True:
        time.sleep(1)
        if stop_flag:
            print("Stop flag set. Stopping script.")
            break


def signal_handler(sig, frame):
    global stop_flag
    print("Signal received. Stopping script.")
    stop_flag = True


def print_resource():
    global stop_flag
    while True:
        time.sleep(2)
        if stop_flag:
            break
        print(f"Current value of shared resource: {shared_resource}")


def main():
    global stop_flag


    signal.signal(signal.SIGINT, signal_handler)


    stop_thread = threading.Thread(target=stop_monitor)
    stop_thread.start()


    print_thread = threading.Thread(target=print_resource)
    print_thread.start()


    num_threads = 5
    threads = []
    for i in range(num_threads):
        worker_thread = threading.Thread(target=worker, args=(i+1,))
        worker_thread.start()
        threads.append(worker_thread)


    for thread in threads:
        thread.join()
        if stop_flag:
            break


    print(f"Final value of shared resource: {shared_resource}")


    stop_thread.join()


    print_thread.join()


if __name__ == "__main__":
    main()
```
