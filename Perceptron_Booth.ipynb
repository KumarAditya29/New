```python
def booth_multiplication(m, q):
    m = int(m)
    q = int(q)
    m_neg = -m
    q_bit = 0
    n = max(m.bit_length(), q.bit_length())
    a = 0
    q_temp = q
    print(f"Multiplicand (m): {m}, Multiplier (q): {q}")
    print(f"Initial A: {a}, Q: {q}, Q-1: {q_bit}")
    for _ in range(n):
        print(f"\nIteration {_ + 1}:")
        print(f"A: {bin(a)[2:].zfill(n)}, Q: {bin(q)[2:].zfill(n)}, Q-1: {q_bit}")
        if (q & 1) == 1 and q_bit == 0:
            a += m
            print(f"Subtracting m: A = A - m => A: {a}")
        elif (q & 1) == 0 and q_bit == 1:
            a += m_neg
            print(f"Adding m: A = A + m_neg => A: {a}")
        a, q = (a << 1) | (q >> 1), (q >> 1)
        q_bit = a & 1
        a >>= 1
    result = (a << n) | q
    print(f"\nFinal A: {a}, Final Q: {q}")
    print(f"Result: {result} (Binary: {bin(result)[2:]})")
    return result
def test_booth_multiplication():
    test_cases = [
        (3, 4),
        (-3, 4),
        (3, -4),
        (-3, -4),
        (5, 5)
    ]
    for m, q in test_cases:
        print(f"\nMultiplying {m} and {q}:")
        booth_multiplication(m, q)
test_booth_multiplication()
```
